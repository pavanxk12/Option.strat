```python
import pandas as pd
import numpy as np

futures = pd.read_csv('/content/drive/MyDrive/Dhaniti/BANKNIFTY_2017_FUTURES.csv')
```

```python
columnstocopy=['Date','Time','Close','Contract','Expiry']
res1=futures[columnstocopy].copy()
```

```python
res1['Time']=pd.to_datetime(res1['Time'])
```

```python
import datetime
desired_time=datetime.time(10,29,59)
```

```python
res2=res1[res1['Time'].dt.time==desired_time]
res2=res2[res2['Contract']=='I']
res2.head()
```

```python
def roundoff(df_row):
  x = df_row['Close']
  if (x % 100 < 50):
    y = x - x % 100
  elif (x%100>50):
    y=x-x%50
  else:
    y=x
  return y
```

```python
value=res2.apply(roundoff,axis=1)
```

```python
res2.insert(4,"ATM",value,True)
```

```python
def roundoff1(df_row):
  x = 1.02*df_row['Close']
  if (x % 100 < 50):
    y = x - x % 100
  elif (x%100>50):
    y=x-x%50
  else:
    y=x
  return y
```

```python
value1=res2.apply(roundoff1,axis=1)
```

```python
res2.insert(4,"OTMCE",value1,True)
```

```python
def roundoff2(df_row):
  x = 0.98*df_row['Close']
  if (x % 100 < 50):
    y = x - x % 100
  elif (x%100>50):
    y=x-x%50
  else:
    y=x
  return y
```

```python
value2=res2.apply(roundoff2,axis=1)
```

```python
res2.insert(4,"OTMPE",value2,True)
```

```python
options['Datetime'] = pd.to_datetime(options['Date'].astype(str) + ' ' + options['Time'].astype(str))
```

```python
threshold_datetime = pd.to_datetime('2017-01-03 10:29:59')

filtered_df = options[options['Datetime'] > threshold_datetime]

# filtered_df = filtered_df.drop(columns=['Datetime'])

filtered_df.head()
```

```python
ticker_columns = ['Ticker_atmce', 'Ticker_atmpe', 'Ticker_otmce', 'Ticker_otmpe']

ticker_variables = {}

for col in ticker_columns:
    ticker_variables[col] = result_df[col].fillna('').tolist()
```

```python
import pandas as pd
import numpy as np

desired_time = pd.to_datetime('2017-01-03 10:29:59')
sum_closex = result_df['sumclosex'].iloc[0]
total_close_sums = []


max_length = max(filtered_df.groupby('Ticker')['Close'].count())

for i in range(max_length):
    close_values_sum = 0

    for col in ticker_columns:
        key = ticker_variables[col][0]
        a = filtered_df[filtered_df['Ticker'] == key]

        if not a.empty and i < len(a):
            close_values_sum += a['Close'].iloc[i]
            # print("Close value for {} after {} at index {}: {}".format(key, desired_time, i, a['Close'].iloc[i]))

    if close_values_sum == 0:
        break

    print("\nSum: ", close_values_sum)
    total_close_sums.append(close_values_sum)

    if close_values_sum <= 0.2 * sum_closex:
        print("Yes")
        break
    elif close_values_sum > 1.3 * sum_closex:
        print("No")
        break

# total_close_sums_array = np.array(total_close_sums)

# print("\nTotal Close Sums Array after {}: {}".format(desired_time, total_close_sums_array))
```
